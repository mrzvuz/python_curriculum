# Scatter

## Explanation

Scatter plots show relationships between two numeric variables.


## Example

Run the code cell below and then change the values.

```python
import matplotlib.pyplot as plt

heights = [160, 170, 180, 190]
weights = [55, 65, 80, 90]

plt.scatter(heights, weights)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```
