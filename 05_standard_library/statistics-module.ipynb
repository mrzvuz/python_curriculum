# Statistics Module

## Explanation

The statistics module calculates mean, median, mode, and other statistics.


## Example

Run the code cell below and then change the values.

```python
import statistics

scores = [80, 90, 75, 90, 100]
print(statistics.mean(scores))
print(statistics.median(scores))
print(statistics.mode(scores))
```
