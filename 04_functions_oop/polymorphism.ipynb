# Polymorphism

## Explanation

Polymorphism means different objects can use the same method name but behave differently.


## Example

Run the code cell below and then change the values.

```python
class Email:
    def send(self):
        print("Sending email")

class SMS:
    def send(self):
        print("Sending SMS")

messages = [Email(), SMS()]

for message in messages:
    message.send()
```

## Exercises

1. Create three shape classes with an area() method.
2. Loop through them and call area().

```python
# Write your solution here.
```

## Possible Solution

```python
# Possible solution
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Square:
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

shapes = [Rectangle(4, 5), Square(3), Circle(2)]
for shape in shapes:
    print(shape.area())
```
