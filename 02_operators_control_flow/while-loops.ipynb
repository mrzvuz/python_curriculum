# While Loops

## Explanation

A while loop repeats while a condition is True. Be careful to change something inside the loop so it eventually stops.


## Example

Run the code cell below and then change the values.

```python
count = 1

while count <= 5:
    print("Count:", count)
    count += 1
```

## Possible Solution

```python
# Possible solution
number = 1
while number <= 10:
    print(number)
    number += 1

countdown = 5
while countdown >= 1:
    print(countdown)
    countdown -= 1
```
