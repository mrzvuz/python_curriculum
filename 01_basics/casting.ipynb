# Casting

## Explanation

Casting converts one data type into another. Common casting functions are int(), float(), str(), list(), tuple(), and set().


## Example

Run the code cell below and then change the values.

```python
number_text = "42"
number = int(number_text)
price = float("19.99")
message = str(100)

print(number + 8)
print(price * 2)
print("Score: " + message)
```
