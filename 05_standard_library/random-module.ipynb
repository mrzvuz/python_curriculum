# Random Module

## Explanation

The random module creates random numbers and random choices.


## Example

Run the code cell below and then change the values.

```python
import random

print(random.randint(1, 6))
print(random.choice(["red", "green", "blue"]))
print(random.sample(range(1, 50), 6))
```
