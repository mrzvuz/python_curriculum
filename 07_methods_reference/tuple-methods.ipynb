# Tuple Methods

## Explanation

Tuples are immutable, so they have fewer methods. The main methods are count() and index().


## Example

Run the code cell below and then change the values.

```python
values = ("a", "b", "a", "c")

print(values.count("a"))
print(values.index("c"))
```
