# Class Methods

## Explanation

A class method receives the class as cls. It is often used for alternate constructors or class-level behavior.


## Example

Run the code cell below and then change the values.

```python
class User:
    user_count = 0

    def __init__(self, name):
        self.name = name
        User.user_count += 1

    @classmethod
    def total_users(cls):
        return cls.user_count

User("Adam")
User("Sara")
print(User.total_users())
```
