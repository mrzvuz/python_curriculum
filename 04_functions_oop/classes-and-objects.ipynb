# Classes and Objects

## Explanation

A class is a blueprint. An object is a real instance created from that blueprint.


## Example

Run the code cell below and then change the values.

```python
class Dog:
    def bark(self):
        print("Woof")

my_dog = Dog()
my_dog.bark()
```
