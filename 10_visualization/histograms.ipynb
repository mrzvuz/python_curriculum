# Histograms

## Explanation

Histograms show the distribution of numeric data.


## Example

Run the code cell below and then change the values.

```python
import matplotlib.pyplot as plt

ages = [18, 20, 21, 21, 22, 25, 30, 31, 35, 40]

plt.hist(ages, bins=5)
plt.show()
```
