# Built-in Functions

## Explanation

Built-in functions are always available in Python. Examples include print(), len(), type(), sum(), min(), max(), sorted(), input(), and range().


## Example

Run the code cell below and then change the values.

```python
numbers = [10, 4, 7, 20]

print(len(numbers))
print(sum(numbers))
print(min(numbers))
print(max(numbers))
print(sorted(numbers))
```
