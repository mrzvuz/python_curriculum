# Tuples

## Explanation

A tuple is an ordered, unchangeable collection. Tuples are useful for values that should stay fixed.


## Example

Run the code cell below and then change the values.

```python
coordinates = (10, 20)
name, age, city = ("Adam", 25, "Istanbul")

print(coordinates)
print(name, age, city)
```
