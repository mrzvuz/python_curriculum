# Keywords

## Explanation

Keywords are reserved words in Python. You cannot use them as variable names.


## Example

Run the code cell below and then change the values.

```python
import keyword

print(keyword.kwlist)
print(keyword.iskeyword('class'))
print(keyword.iskeyword('student'))
```
