# Set Methods

## Explanation

Set methods support unique values and math-like operations such as union, intersection, and difference.


## Example

Run the code cell below and then change the values.

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```
