# Plotting

## Explanation

Plotting turns data into visual form. Start with simple line charts before moving to specialized charts.


## Example

Run the code cell below and then change the values.

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar']
sales = [100, 150, 120]

plt.plot(months, sales)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
