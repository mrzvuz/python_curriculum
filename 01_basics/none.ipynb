# None

## Explanation

None means no value. It is useful when a variable exists, but does not have a meaningful value yet.


## Example

Run the code cell below and then change the values.

```python
selected_user = None

if selected_user is None:
    print("No user selected yet")
else:
    print(selected_user)
```
