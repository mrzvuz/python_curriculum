# If Else

## Explanation

if, elif, and else let your program choose different paths based on conditions.


## Example

Run the code cell below and then change the values.

```python
score = 82

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("Needs practice")
```
