# Comments

## Explanation

Comments are notes for humans. Python ignores comments when running code. Use comments to explain why something exists, not every tiny step.


## Example

Run the code cell below and then change the values.

```python
# This is a single-line comment
name = "Adam"  # This comment explains this variable

"""
This is a multi-line string.
It is often used as a longer note or docstring.
"""
print(name)
```
