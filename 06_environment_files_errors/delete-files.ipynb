# Delete Files

## Explanation

Files can be deleted with pathlib's unlink() or os.remove(). Always check that the file exists before deleting.


## Example

Run the code cell below and then change the values.

```python
from pathlib import Path

path = Path("temporary_file.txt")
path.write_text("Delete me", encoding="utf-8")

if path.exists():
    path.unlink()
    print("File deleted")
```
