# Inner Classes

## Explanation

An inner class is a class defined inside another class. It can be useful when one object only makes sense inside another.


## Example

Run the code cell below and then change the values.

```python
class Computer:
    class CPU:
        def info(self):
            return "CPU is processing"

    def __init__(self, brand):
        self.brand = brand
        self.cpu = Computer.CPU()

computer = Computer("Lenovo")
print(computer.brand)
print(computer.cpu.info())
```

## Possible Solution

```python
# Possible solution
class House:
    class Room:
        def __init__(self, name):
            self.name = name

        def describe(self):
            return f"This is the {self.name}."

    def __init__(self):
        self.room = House.Room("kitchen")

house = House()
print(house.room.describe())
```
