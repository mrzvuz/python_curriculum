# Subplot

## Explanation

Subplots let you show multiple charts in one figure.


## Example

Run the code cell below and then change the values.

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title('Squares')

plt.subplot(1, 2, 2)
plt.plot([1, 2, 3], [1, 8, 27])
plt.title('Cubes')

plt.tight_layout()
plt.show()
```
