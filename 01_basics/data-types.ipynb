# Data Types

## Explanation

A data type tells Python what kind of value something is. Common types include str, int, float, bool, list, tuple, set, dict, and NoneType.


## Example

Run the code cell below and then change the values.

```python
values = ["Python", 42, 3.14, True, ["a", "b"], ("x", "y"), {"red", "blue"}, {"name": "Adam"}, None]

for value in values:
    print(value, "=>", type(value))
```
