# Markers

## Explanation

Markers show individual data points on a plot.


## Example

Run the code cell below and then change the values.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [10, 20, 15, 30], marker='o')
plt.show()
```
