# Match

## Explanation

match is Python's pattern matching statement. It is useful when you want clean alternatives for one value. It requires Python 3.10 or newer.


## Example

Run the code cell below and then change the values.

```python
command = "start"

match command:
    case "start":
        print("Starting")
    case "stop":
        print("Stopping")
    case "pause":
        print("Pausing")
    case _:
        print("Unknown command")
```
