# List Methods

## Explanation

List methods modify or inspect lists. Common methods include append(), extend(), insert(), remove(), pop(), sort(), and reverse().


## Example

Run the code cell below and then change the values.

```python
numbers = [3, 1, 2]
numbers.append(4)
numbers.sort()
print(numbers)

removed = numbers.pop()
print(removed)
print(numbers)
```
