## 1. For Loops

A `for` loop repeats code for each item in a collection.

Think of it like this:

`for each item in this group, do this action`

The variable after `for` is temporary. It changes automatically each time the loop runs.

```python
students = ["Adam", "Sara", "Mehmet", "Lina"]

for student in students:
    print("Hello", student)
```

What Python does:

1. First loop: `student = "Adam"`
2. Second loop: `student = "Sara"`
3. Third loop: `student = "Mehmet"`
4. Fourth loop: `student = "Lina"`

You do not manually change `student`. Python changes it for you.

## For Loop Example With 4 Values Per Item

Each product below has 4 pieces of information:

1. product name
2. category
3. price
4. stock quantity

This is a good way to understand how loops work with structured data.

```python
products = [
    ["Laptop", "Electronics", 1200, 5],
    ["Mouse", "Electronics", 25, 50],
    ["Notebook", "Stationery", 4, 100],
    ["Chair", "Furniture", 80, 10],
]

for product in products:
    name = product[0]
    category = product[1]
    price = product[2]
    stock = product[3]
    total_value = price * stock

    print(name, "|", category, "|", "Total value:", "${:.2f}".format(total_value))
```

Inside the loop, `product` becomes one list at a time.

First time:

`product = ["Laptop", "Electronics", 1200, 5]`

Second time:

`product = ["Mouse", "Electronics", 25, 50]`

The loop is simply repeating the same logic for every product.

## 2. Functions

A function is reusable code.

Think of a function like a small machine:

1. You give it inputs.
2. It does work.
3. It can return an output.

The names inside the parentheses are called **parameters**.

## Function With 3 Parameters

This function receives:

1. first name
2. last name
3. city

```python
def introduce_person(first_name, last_name, city):
    message = f"Hello, my name is {first_name} {last_name}. I live in {city}."
    return message

result = introduce_person("Adam", "Smith", "Istanbul")
print(result)
```

`first_name`, `last_name`, and `city` are parameters.

`"Adam"`, `"Smith"`, and `"Istanbul"` are arguments.

Parameters are placeholders. Arguments are the real values.

## Function With 4 Parameters

This function calculates the total price of an order.

It receives:

1. product name
2. quantity
3. price per item
4. tax rate

```python
def calculate_order_total(product_name, quantity, price_each, tax_rate):
    subtotal = quantity * price_each
    tax = subtotal * tax_rate
    total = subtotal + tax

    return f"{product_name} total: {total}"

print(calculate_order_total("Notebook", 3, 15, 0.20))
print(calculate_order_total("Pen", 10, 2, 0.20))
```

Why functions are useful:

Without the function, you would repeat the same calculation again and again.

With the function, you write the calculation once and reuse it with different values.

## Function With 5 Parameters

This function creates a student report.

It receives:

1. student name
2. math score
3. English score
4. science score
5. attendance status

```python
def create_student_report(name, math_score, english_score, science_score, attendance):
    average = (math_score + english_score + science_score) / 3

    report = {
        "name": name,
        "math": math_score,
        "english": english_score,
        "science": science_score,
        "average": average,
        "attendance": attendance,
    }

    return report

report_one = create_student_report("Sara", 90, 85, 95, "Excellent")
report_two = create_student_report("Mehmet", 70, 75, 80, "Good")

print(report_one)
print(report_two)
```

The most important function pattern:

`define once, use many times`

If you copy and paste the same code many times, it might be a good function.

## 3. Classes

A class is a blueprint for creating objects.

A class groups together:

1. data
2. actions

Example:

A student has data: name, age, city, course, grade.

A student can do actions: introduce themselves, check if passing.

## Class With 5 Parameters

`__init__` runs automatically when you create a new object.

`self` means: this specific object.

```python
class Student:
    def __init__(self, name, age, city, course, grade):
        self.name = name
        self.age = age
        self.city = city
        self.course = course
        self.grade = grade

    def introduce(self):
        return f"My name is {self.name}. I am learning {self.course}."

    def is_passing(self):
        return self.grade >= 50

student_one = Student("Adam", 25, "Istanbul", "Python", 88)
student_two = Student("Lina", 22, "Ankara", "Data Analysis", 45)

print(student_one.introduce())
print(student_one.is_passing())

print(student_two.introduce())
print(student_two.is_passing())
```

Important class vocabulary:

- `Student` is the class.
- `student_one` is an object.
- `student_two` is another object.
- `name`, `age`, `city`, `course`, and `grade` are object data.
- `introduce()` and `is_passing()` are methods.

A method is just a function inside a class.

## Class With 4 Parameters

This class represents a bank account.

It receives:

1. owner
2. bank name
3. account type
4. balance

```python
class BankAccount:
    def __init__(self, owner, bank_name, account_type, balance):
        self.owner = owner
        self.bank_name = bank_name
        self.account_type = account_type
        self.balance = balance

    def deposit(self, amount):
        self.balance = self.balance + amount
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            return "Not enough money"

        self.balance = self.balance - amount
        return self.balance

    def summary(self):
        return f"{self.owner} has {self.balance} in a {self.account_type} account."

account = BankAccount("Adam", "Python Bank", "Savings", 1000)

print(account.summary())
print(account.deposit(250))
print(account.withdraw(400))
print(account.summary())
```

Notice the difference:

A function usually works with values you pass into it.

A class creates objects that remember their own values over time.

The bank account remembers its balance after deposit and withdrawal.

## Putting All Three Together

Now we will use:

1. a class to represent products
2. a function to print a product summary
3. a for loop to repeat the summary for every product

```python
class Product:
    def __init__(self, name, category, price, stock):
        self.name = name
        self.category = category
        self.price = price
        self.stock = stock

    def total_stock_value(self):
        return self.price * self.stock


def print_product_summary(name, category, price, stock, total_value):
    print(f"{name} | {category} | price: {price} | stock: {stock} | value: {total_value}")


products = [
    Product("Laptop", "Electronics", 1200, 5),
    Product("Mouse", "Electronics", 25, 50),
    Product("Notebook", "Stationery", 4, 100),
]

for product in products:
    print_product_summary(
        product.name,
        product.category,
        product.price,
        product.stock,
        product.total_stock_value(),
    )
```

What happened?

1. `Product` is a class.
2. `Product(...)` creates product objects.
3. `products` is a list of product objects.
4. The `for` loop goes through one product at a time.
5. The function receives 5 values and prints a clean summary.

This is a very common Python pattern.

## Quick Memory Guide

`for` loop: use it when you want to repeat an action for every item.

`function`: use it when you want reusable code that receives inputs.

`class`: use it when you want to model a real thing that has data and actions.

A simple way to remember:

- loop = repeat
- function = reusable action
- class = blueprint for objects
