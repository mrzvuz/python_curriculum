# Class Properties

## Explanation

Class properties belong to the class. Instance properties belong to each object.


## Example

Run the code cell below and then change the values.

```python
class Employee:
    company = "Python Corp"

    def __init__(self, name, role):
        self.name = name
        self.role = role

employee = Employee("Adam", "Developer")
print(employee.name)
print(employee.company)
```

## Possible Solution

```python
# Possible solution
class Student:
    school = "Python Academy"

    def __init__(self, name):
        self.name = name

student_one = Student("Adam")
student_two = Student("Sara")

print(student_one.name, student_one.school)
print(student_two.name, student_two.school)
```
