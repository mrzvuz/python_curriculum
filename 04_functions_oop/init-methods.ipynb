# __init__ Methods

## Explanation

__init__ runs automatically when a new object is created. Use it to set starting values.


## Example

Run the code cell below and then change the values.

```python
class Student:
    def __init__(self, name, age, course, grade):
        self.name = name
        self.age = age
        self.course = course
        self.grade = grade

student = Student("Adam", 25, "Python", 95)
print(student.name, student.grade)
```

## Possible Solution

```python
# Possible solution
class Product:
    def __init__(self, name, category, price, stock):
        self.name = name
        self.category = category
        self.price = price
        self.stock = stock

product = Product("Laptop", "Electronics", 1200, 5)
print(product.name)
print(product.category)
print(product.price)
print(product.stock)
```
