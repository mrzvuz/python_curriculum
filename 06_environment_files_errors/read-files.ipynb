# Read Files

## Explanation

Reading a file loads existing content from disk. pathlib makes simple file reading clear.


## Example

Run the code cell below and then change the values.

```python
from pathlib import Path

path = Path("sample_read.txt")
path.write_text("Line 1\nLine 2", encoding="utf-8")

for line in path.read_text(encoding="utf-8").splitlines():
    print(line)
```

## Possible Solution

```python
# Possible solution
from pathlib import Path

path = Path("lines.txt")
path.write_text("one\ntwo\nthree", encoding="utf-8")
lines = path.read_text(encoding="utf-8").splitlines()

for line in lines:
    print(line)

print("Line count:", len(lines))
```
