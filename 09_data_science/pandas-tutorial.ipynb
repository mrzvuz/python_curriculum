# Pandas Tutorial

## Explanation

Pandas is used for tabular data. DataFrames are like powerful spreadsheets in Python.


## Example

Run the code cell below and then change the values.

```python
import pandas as pd

data = {
    "name": ["Adam", "Sara", "Lina"],
    "score": [90, 85, 95],
}

df = pd.DataFrame(data)
print(df)
print(df["score"].mean())
```
