# Lists

## Explanation

A list is an ordered, changeable collection. Lists can contain any data type and allow duplicate values.


## Example

Run the code cell below and then change the values.

```python
colors = ["red", "green", "blue"]
colors.append("yellow")
colors[1] = "black"

print(colors)
print(colors[0])
print(len(colors))
```
