# Add Two Numbers

## Explanation

Adding numbers is simple, but remember that input() returns strings, so cast input values before math.


## Example

Run the code cell below and then change the values.

```python
a = 10
b = 15
print(a + b)

def add_numbers(first, second):
    return first + second

print(add_numbers(5, 7))
```
