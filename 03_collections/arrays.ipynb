# Arrays

## Explanation

Python lists are usually used as arrays. For numeric arrays, NumPy is the common choice. The built-in array module can also store typed arrays.


## Example

Run the code cell below and then change the values.

```python
from array import array

scores = array('i', [80, 90, 75, 100])
scores.append(85)

for score in scores:
    print(score)
```
