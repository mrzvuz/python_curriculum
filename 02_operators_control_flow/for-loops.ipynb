# For Loops

## Explanation

A for loop repeats once for every item in a sequence such as a list, tuple, string, dictionary, or range.


## Example

Run the code cell below and then change the values.

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)

for letter in "Python":
    print(letter)
```
