# Inheritance

## Explanation

Inheritance lets one class reuse another class. The child class gets methods and properties from the parent class.


## Example

Run the code cell below and then change the values.

```python
class Animal:
    def speak(self):
        return "Some sound"

class Cat(Animal):
    def speak(self):
        return "Meow"

cat = Cat()
print(cat.speak())
```
