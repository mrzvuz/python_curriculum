# Iterators

## Explanation

An iterator is an object that gives values one at a time. iter() creates an iterator and next() gets the next value.


## Example

Run the code cell below and then change the values.

```python
numbers = [10, 20, 30]
iterator = iter(numbers)

print(next(iterator))
print(next(iterator))
print(next(iterator))
```
