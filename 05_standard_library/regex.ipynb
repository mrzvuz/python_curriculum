# RegEx

## Explanation

Regular expressions search for text patterns. Python uses the re module for RegEx.


## Example

Run the code cell below and then change the values.

```python
import re

text = "Email me at adam@example.com"
match = re.search(r"\w+@\w+\.\w+", text)

if match:
    print(match.group())
```
