# Try Except Error Handlers

## Explanation

try and except let your program handle errors instead of crashing immediately.


## Example

Run the code cell below and then change the values.

```python
try:
    number = int("abc")
except ValueError:
    print("That text cannot become an integer")
finally:
    print("This always runs")
```

## Possible Solution

```python
# Possible solution
try:
    result = 10 / 0
except ZeroDivisionError:
    print("You cannot divide by zero.")

student = {"name": "Adam"}
try:
    print(student["grade"])
except KeyError:
    print("The grade key does not exist.")
```
