# Dates

## Explanation

The datetime module helps you work with dates and times.


## Example

Run the code cell below and then change the values.

```python
from datetime import datetime, date, timedelta

now = datetime.now()
today = date.today()
tomorrow = today + timedelta(days=1)

print(now)
print(today)
print(tomorrow)
```
