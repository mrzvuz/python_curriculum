# Dictionary Methods

## Explanation

Dictionary methods help access and update key-value data. Common methods include get(), keys(), values(), items(), update(), and pop().


## Example

Run the code cell below and then change the values.

```python
person = {"name": "Adam", "age": 25}

print(person.get("name"))
print(person.keys())
print(person.values())
person.update({"city": "Istanbul"})
print(person)
```
