### Challenge 7.2: Remove Duplicates from a Playlist

**Scenario:** You have a playlist with duplicate songs.

**Your Task:**
1. Playlist: ["Song A", "Song B", "Song A", "Song C", "Song B", "Song D"]
2. Remove duplicates while keeping order
3. Count how many duplicates were removed

**Try it yourself:**

```python
# YOUR CODE HERE
```

**Solution:**

```python
# Solution
playlist = ["Song A", "Song B", "Song A", "Song C", "Song B", "Song D"]

# Remove duplicates while keeping order
seen = set()
unique_playlist = []
for song in playlist:
    if song not in seen:
        seen.add(song)
        unique_playlist.append(song)

duplicates_removed = len(playlist) - len(unique_playlist)

print(f"Original playlist ({len(playlist)} songs): {playlist}")
print(f"Cleaned playlist ({len(unique_playlist)} songs): {unique_playlist}")
print(f"Duplicates removed: {duplicates_removed}")
```

### Challenge 8.2: Product Inventory System

**Scenario:** Manage a small store's inventory.

**Your Task:**
1. Create inventory with products: name, quantity, price
   - Laptop: 5 units, $999
   - Mouse: 50 units, $25
   - Keyboard: 30 units, $75
2. Calculate total inventory value
3. Find products with low stock (< 10 units)
4. Update a product's quantity
5. Add a new product

**Try it yourself:**

```python
# YOUR CODE HERE
```

**Solution:**

```python
# Solution
inventory = {
    "Laptop": {"quantity": 5, "price": 999},
    "Mouse": {"quantity": 50, "price": 25},
    "Keyboard": {"quantity": 30, "price": 75}
}

# Calculate total inventory value
total_value = sum(item['quantity'] * item['price'] for item in inventory.values())
print(f"Total Inventory Value: ${total_value:,.2f}")

# Find low stock items
low_stock = {name: info for name, info in inventory.items() if info['quantity'] < 10}
print(f"\nLow Stock Items (< 10): {low_stock}")

# Update product
inventory["Mouse"]["quantity"] = 45
print(f"\nUpdated Mouse quantity to 45")

# Add new product
inventory["Monitor"] = {"quantity": 8, "price": 299}
print(f"Added Monitor to inventory")

# Show final inventory
print(f"\nFinal Inventory:")
for product, info in inventory.items():
    value = info['quantity'] * info['price']
    print(f"  {product}: {info['quantity']} units @ ${info['price']} = ${value:,.2f}")
```

### Challenge 9.2: Weather-Based Activity Recommender

**Scenario:** Recommend activities based on weather conditions.

**Your Task:**
1. Given: temperature, weather condition (sunny/rainy/cloudy), is_weekend
2. Recommend activity based on conditions
3. Provide reasoning

**Rules:**
- Sunny & warm (>75°F) & weekend → Beach or Hiking
- Rainy → Indoor movie or read
- Cloudy & cool → Coffee shop or library
- Cold (<45°F) → Hot chocolate and cozy indoors

**Try it yourself:**

```python
# YOUR CODE HERE
temperature = 78
weather = "sunny"
is_weekend = True
```

**Solution:**

```python
# Solution
temperature = 78
weather = "sunny"
is_weekend = True

print(f"Weather: {temperature}°F, {weather.capitalize()}, {'Weekend' if is_weekend else 'Weekday'}")
print("-" * 40)

if weather == "rainy":
    activity = "Watch a movie or read a book"
    reason = "Perfect weather for indoor activities"
elif temperature < 45:
    activity = "Stay cozy indoors with hot chocolate"
    reason = "It's too cold to be outside"
elif weather == "sunny" and temperature > 75 and is_weekend:
    activity = "Go to the beach or hiking"
    reason = "Beautiful weather and you have free time!"
elif weather == "sunny" and temperature > 75:
    activity = "Go for a walk or picnic"
    reason = "Great weather, perfect for outdoor activities"
elif weather == "cloudy":
    activity = "Visit a coffee shop or library"
    reason = "Good day for quiet, relaxed activities"
else:
    activity = "Flexible - choose what you enjoy"
    reason = "Weather is moderate"

print(f"Recommendation: {activity}")
print(f"Why: {reason}")
```
