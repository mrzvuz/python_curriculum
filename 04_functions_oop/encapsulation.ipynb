# Encapsulation

## Explanation

Encapsulation protects internal data and exposes controlled methods. Python uses naming conventions like _protected and __private.


## Example

Run the code cell below and then change the values.

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def get_balance(self):
        return self.__balance

account = BankAccount("Adam", 100)
account.deposit(50)
print(account.get_balance())
```

## Possible Solution

```python
# Possible solution
class Wallet:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def add_money(self, amount):
        if amount > 0:
            self.__balance += amount

wallet = Wallet(100)
wallet.add_money(50)
print(wallet.get_balance())
```
