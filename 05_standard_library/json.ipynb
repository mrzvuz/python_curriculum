# JSON

## Explanation

JSON is a common format for sharing structured data. The json module converts between Python objects and JSON strings.


## Example

Run the code cell below and then change the values.

```python
import json

user = {"name": "Adam", "age": 25, "skills": ["Python", "Pandas"]}
json_text = json.dumps(user, indent=2)
print(json_text)

back_to_python = json.loads(json_text)
print(back_to_python["name"])
```
