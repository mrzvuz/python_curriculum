# Operators

## Explanation

Operators perform actions on values. Python has arithmetic, assignment, comparison, logical, identity, membership, and bitwise operators.


## Example

Run the code cell below and then change the values.

```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a > b)
print(a > 5 and b < 5)
```
