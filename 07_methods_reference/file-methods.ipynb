# File Methods

## Explanation

File objects have methods like read(), readline(), readlines(), write(), and close(). The with statement is preferred because it closes automatically.


## Example

Run the code cell below and then change the values.

```python
from pathlib import Path

path = Path("file_methods_demo.txt")
with path.open("w", encoding="utf-8") as file:
    file.write("First line\nSecond line\n")

with path.open("r", encoding="utf-8") as file:
    print(file.readline())
```
