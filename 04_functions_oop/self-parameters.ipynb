# Self Parameters

## Explanation

self refers to the current object. It lets each object keep its own data.


## Example

Run the code cell below and then change the values.

```python
class Counter:
    def __init__(self):
        self.value = 0

    def increase(self):
        self.value += 1

counter_one = Counter()
counter_two = Counter()
counter_one.increase()
counter_one.increase()
counter_two.increase()

print(counter_one.value)
print(counter_two.value)
```

## Possible Solution

```python
# Possible solution
class Score:
    def __init__(self):
        self.points = 0

    def add_point(self):
        self.points += 1

score_one = Score()
score_two = Score()

score_one.add_point()
score_one.add_point()
score_two.add_point()

print(score_one.points)
print(score_two.points)
```
