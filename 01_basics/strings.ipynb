# Strings

## Explanation

A string is text. Strings can be indexed, sliced, joined, formatted, uppercased, lowercased, and searched.


## Example

Run the code cell below and then change the values.

```python
language = "Python"

print(language[0])
print(language[1:4])
print(language.upper())
print(language.lower())
print("I am learning " + language)
```
