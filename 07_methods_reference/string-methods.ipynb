# String Methods

## Explanation

String methods are actions attached to string values, such as upper(), lower(), strip(), replace(), split(), and startswith().


## Example

Run the code cell below and then change the values.

```python
text = "  Python is fun  "

print(text.strip())
print(text.upper())
print(text.replace("fun", "powerful"))
print(text.strip().split())
```
