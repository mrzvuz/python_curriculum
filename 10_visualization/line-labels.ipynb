## Explanation

## Example

Run the code cell below and then change the values.

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [10, 20, 30], label='Sales')
plt.plot([1, 2, 3], [8, 15, 25], label='Profit')
plt.legend()
plt.show()
```
