# Dictionaries

## Explanation

A dictionary stores key-value pairs. Use a key to look up a value.


## Example

Run the code cell below and then change the values.

```python
student = {"name": "Adam", "age": 25, "course": "Python"}
student["grade"] = 95

print(student["name"])
print(student.get("course"))
print(student)
```
