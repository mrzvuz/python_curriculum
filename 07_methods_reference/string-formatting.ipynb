# String Formatting

## Explanation

String formatting inserts values into text. f-strings are the modern, readable choice.


## Example

Run the code cell below and then change the values.

```python
name = "Adam"
age = 25
price = 19.987

print(f"My name is {name} and I am {age}.")
print(f"Price: ${price:.2f}")
```
